```
from collections import defaultdict 
```

```
# 將班次一到五與其所有時間用一種資料型態個別保存

with open('example.csv', newline='') as csvfile:
    # 讀取 CSV 檔案內容
    next(csvfile)
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    d_time = defaultdict(list) 
    for row in rows:
        d_time['row5'].append(row[5])
        d_time['row6'].append(row[6])
        d_time['row7'].append(row[7])
        d_time['row8'].append(row[8])
        d_time['row9'].append(row[9])

        # print([row[5], row[6], row[7], row[8], row[9]])
        # list_time.append([row[5], row[6], row[7], row[8], row[9]])
```
