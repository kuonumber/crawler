```
import xml.etree.ElementTree as ET 

# 存取檔案
tree = ET.parse('./crawler/example/sample.xml') 
root = tree.getroot()

# 存取我們的資訊
print(root[0].text)

# 用迴圈存取我們的資訊
chapters = root[2]
for chapter in chapters:
    print (chapter.attrib['name'], chapter.text)
```
