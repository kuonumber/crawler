```
import requests
```

```
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
response
```

```
import json
data = json.loads(response)
d_c = 0
d_l = 0 
for d in data:
    print(d['title'], d['commentCount'], d['createdAt'], d['likeCount'])
    d_c += d['commentCount']
    d_l += d['likeCount']
```
